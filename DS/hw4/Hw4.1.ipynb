```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
time = np.linspace(0, 11, 12)
```

```python
time
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(speed, color='blue', label='Speed')

plt.xlim([0, 11])
plt.ylim([0, 130])
plt.legend()
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', linestyle=':');
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
f = interp1d(time, speed, kind='cubic')

plt.plot(time, speed, 'o')
more_speed = np.linspace(0, 11, 10000)
plt.plot(more_speed, f(more_speed))
plt.xlim([0, 11])
plt.ylim([0, 130])
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', linestyle=':');
```

```python
result = integrate.quad(f, 0, 11)
print(result)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
f_two = interp1d(time, speed, kind='quadratic')

plt.plot(time, speed, 'o')
more_speed_two = np.linspace(0, 11, 10000)
plt.plot(more_speed_two, f_two(more_speed_two))
plt.xlim([0, 11])
plt.ylim([0, 130])
plt.grid(which='major')
plt.minorticks_on()
plt.grid(which='minor', linestyle=':');
```

```python
result = integrate.quad(f_two, 0, 11)
print(result)
```
