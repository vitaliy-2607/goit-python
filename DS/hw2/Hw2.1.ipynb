```python
import pandas as pd

tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C", decimal=',', thousands='_')
table = tmp[10]
table
```

```python
import numpy as np
table_v1 = table.replace('—', np.nan)
table_v1
```

```python
table_v1.dtypes
```

```python
table_v2 = table_v1.fillna(0)
table_v2
```

```python
table_v2.isnull().sum()
```

```python
table_v2.drop(27)
```

```python
table_line = table_v2.astype({'1990': float, '2000': float, '2012': float, '2014': float, '2019': float})
table_line
table_column = table_v1.fillna(table_line.mean().round(1))
table_column = table_column.astype({'1990': float, '2000': float, '2012': float, '2014': float, '2019': float})
table_column
```
